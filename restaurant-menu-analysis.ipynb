# 4. Analysis and Data Visualization

```python
#the most popular menu items based on the popularity score
MenuItems_byPopularity = df.groupby(['Menu Item'])['Popularity'].sum().reset_index()
MenuItems_byPopularity
```

```python
fig = px.treemap(MenuItems_byPopularity ,path=['Menu Item'], 
                 values='Popularity',  
                 color='Popularity',  
                 color_continuous_scale='RdBu')
fig.update_layout(margin = dict(t=50, l=25, r=25, b=25))
fig.show()
```

```python
#items in each category 
ItemsInCategory = df.groupby(['Category'])['Menu Item'].count().reset_index()
ItemsInCategory
```

```python
#the percentage of items in each category
fig = px.pie(ItemsInCategory , values = 'Menu Item' , names = 'Category'  , title = 'Percentage of items in each category ')
fig.show()
```

```python
#the average price for each category
AvgPriceCategory = df.groupby(['Category'])['Price'].mean().reset_index()
AvgPriceCategory
```

```python
fig = px.bar(AvgPriceCategory, x = 'Category', y = 'Price', height = 400,
             title = 'the average price for each category')
fig.show()
```

```python
#the average popularity for each category
AvgpopularityCategory = df.groupby(['Category'])['Popularity'].mean().reset_index()
AvgpopularityCategory
```

```python
fig = px.bar(AvgpopularityCategory, x = 'Category', y = 'Popularity', height = 400,
             title = 'the average popularity for each category')
fig.show()
```
